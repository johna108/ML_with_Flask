```python
import pandas as pd

# Sample DataFrame (replace with your actual DataFrame)
data = pd.DataFrame({
    'Brand': ['Maruti', 'Hyundai', 'Hyundai', 'Maruti'],
    'Year': [2012, 2015, 2020, 2012],
    'City': ['Delhi', 'Delhi', 'Mumbai', 'Delhi'],
    'Kms_Driven': [45000, 35000, 15000, 60000],
    'Customer_Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Discount_Amount': [5000, 7000, 6000, 8000]
})
```

```python

# Filter DataFrame
filtered_data = data.loc[
    (data['City'] == 'Delhi') &
    (data['Kms_Driven'] > 30000)
]

print("Filtered Data:")
print(filtered_data)

# Sample DataFrame for join operations
other_data = pd.DataFrame({
    'Brand': ['Maruti', 'Hyundai'],
    'Customer_Name': ['Alice', 'Bob'],
    'Discount_Amount': [5000, 7000]
})

```

```python
#Inner Join
inner_join = pd.merge(filtered_data, other_data, on=['Brand', 'Customer_Name', 'Discount_Amount'], how='inner')
print("\nInner Join Result:")
print(inner_join)

#Outer Join
outer_join = pd.merge(filtered_data, other_data, on=['Brand', 'Customer_Name', 'Discount_Amount'], how='outer')
print("\nOuter Join Result:")
print(outer_join)

#Left Join
left_join = pd.merge(filtered_data, other_data, on=['Brand', 'Customer_Name', 'Discount_Amount'], how='left')
print("\nLeft Join Result:")
print(left_join)

#Right Join
right_join = pd.merge(filtered_data, other_data, on=['Brand', 'Customer_Name', 'Discount_Amount'], how='right')
print("\nRight Join Result:")
print(right_join)
```
