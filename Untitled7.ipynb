```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

# Sample data
data = {'feature1': [10, 20, 30, 40, 50],
        'feature2': [100, 200, 300, 400, 500],
        'feature3':[10,144,121,432,423]}
df = pd.DataFrame(data)
print("original dataframe")
print(df)

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Apply Label Encoding to 'feature2'
df['feature2'] = label_encoder.fit_transform(df['feature2'])

# Apply Label Encoding to 'feature3'
df['feature3'] = label_encoder.fit_transform(df['feature3'])

print("\nDataFrame after label encoding")
print(df)

# Initialize OneHotEncoder (using sparse=False)
one_hot_encoder = OneHotEncoder()

# Select the columns to encode
columns_to_encode = ['feature1', 'feature2']

# Fit and transform the data
encoded_data = one_hot_encoder.fit_transform(df[columns_to_encode])

# Create a new dataframe with the encoded data
encoded_columns = one_hot_encoder.get_feature_names_out(columns_to_encode)
encoded_df = pd.DataFrame(encoded_data, columns=encoded_columns)

print("\nDataFrame after one-hot encoding")
print(encoded_df)
```
