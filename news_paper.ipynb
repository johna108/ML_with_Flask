```python
import statsmodels.formula.api as smf
model = smf.ols('sunday ~ daily', data=df).fit()

sns.regplot(x='daily', y = 'sunday', data = df)
```

```python
#t and p-Values
print(model.tvalues, '\n')
print(model.pvalues)
```

```python
#R squared values
(model.rsquared, model.rsquared_adj)
```
