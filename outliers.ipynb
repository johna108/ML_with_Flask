```python
import numpy as np

#sample data(simple list of data points)
data = [10,12,14,13,100,87,5,15,16,14,13,11]

#step:1 Calculate(25Th percentile) and Q3 (&5th percentile)
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)

#Step 2: Calculate IQR( Interquartile Range)
IQR = Q3 = Q1

#step3: Determine the lower and upper bondsa for outliers
Lower_bound = Q1 - 1.5 * IQR
Upper_bound = Q3 + 1.5 * IQR

#step4: Find the Outlier(Values outside the lower and upper bounds)
outliers = [x for x in data if x < Lower_bound or x > Upper_bound]

print(f"Q1 (25th percentile):{Q1}")
print(f"Q3 (75th percentile):{Q3}")
print(f"IQR:{IQR}")
print(f"Lower Bound:{Lower_bound}")
print(f"Upper Bound:{Upper_bound}")
print(f"Outliers:{outliers}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Create a sample dataset with some outliers
np.random.seed(10)
data = pd.DataFrame({
    'value': np.concatenate([np.random.normal(0,1,100),np.random.normal(10,10,10)])

})

#calculate the IQR for box plot-based outlier detection
Q1 = data['value'].quantile(0.25)
Q3 = data['value'].quantile(0.75)
IQR = Q3 - Q1

#define outlier threshods
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

#identify outliers
outliers = data[(data['value'] < lower_bound) | (data['value'] > upper_bound)]
print(f"outliers based on box plot criteria:\n{outliers}")

#Create a box plot and violin plot to visualize the data and outliers
plt.figure(figsize=(12,6))

#box plot
plt.figures=(10,5)
plt.subplot(1,2,1)
sns.boxplot(x=data['value'])
plt.title('Box Plot')

#violin plot with box plot inside for comparison
plt.subplot(1,2,2)
sns.violinplot(x=data['value'])
plt.title('Violin Plot')

plt.tight_layout()
plt.show()
```
