```python
import statistics
import matplotlib.pyplot as plt

#sample data
temperature = [25,28,32,35,38,40,42,45,30,33,36,37]
ice_cream_sales = [200,250,300,350,400,420,430,450,280,310,360,370]
advertising_budget = [500,600,700,800,900,1000,1100,1200,550,650,750,850]
customer_count = [100,150,200,250,300,350,400,450,180,230,280,330]

#scatter plot
plt.figure(figsize=(8,6))
plt.scatter(temperature,ice_cream_sales,advertising_budget,customer_count)
plt.xlabel('Temperature')
plt.ylabel('Ice Cream Sales')
plt.title('Temperature vs Ice Cream Sales')
plt.show()
```
