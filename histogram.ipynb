```python
import matplotlib.pyplot as plt

data = [1,2,3,4,10,10,10,18]
plt.hist(data, bins = 10, edgecolor= 'black')

#set the title and labels for plot
plt.title('Simple Histogram Example')
plt.xlabel('Numbers')
plt.ylabel('Count')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(42)
data = np.random.randint(7,67,1213)
plt.hist(data, bins = 10, edgecolor= 'Black', color = "Red")

#set the title and labels for plot
plt.title("Hitsogram for cancer patients")
plt.xlabel('Age')
plt.ylabel('Number of Patients')

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'semester': ['sem1', 'sem1', 'sem1', 'sem1', 'sem1', 'sem2', 'sem2', 'sem2', 'sem2', 'sem2'],
    'hours_studied': [5, 8, 4, 5, 6, 2, 3, 4, 5, 6]
}

df = pd.DataFrame(data)

# Create a box plot
sns.boxplot(x='semester', y='hours_studied', data=df)

plt.title('Box Plot of Hours Studied by Semester')
plt.xlabel('Semester')
plt.ylabel('Hours Studied')

plt.show()
```
